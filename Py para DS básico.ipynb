```
idades = [18, 22, 15, 50]

for i in idades:
  if i < 18:
    print('Você não pode dirigir')
  else:
    print('Você pode dirigir')
```

```
def verificar_dirigir(idade):
  if idade < 18:
    print(f'{idade} anos de idade, NAO TEM permissão para dirigir')
  else:
    print(f'{idade} anos de idade, TEM permissão para dirigir')

for i in idades:
  verificar_dirigir(i)
```
